```python
#THIS CELL CONVERTS THE GIVEN .xlsx WORKBOOK SHEETS INTO pandas DataFrames FOR ANALYSIS

#package imports
import numpy as np
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile

#reading Excel file using pandas
xls = pd.ExcelFile("/Users/BrandonToushan/Desktop/interview.xlsx") #calling path on my computer

#formatting sheets from .xlsx workbook into pandas DataFrames using pandas.read_excel
raw_order = pd.read_excel(xls, 'raw_order') 
raw_customer = pd.read_excel(xls, 'raw_customer') 
raw_customer_meta = pd.read_excel(xls, 'raw_customer_meta')
raw_product = pd.read_excel(xls, 'raw_product')
```

```python
#QUESTION 1: HOW MANY ORDERS WERE CREATED AND WHAT IS THE TOTAL VALUE?

#i)the number of orders created is len of the DataFrame raw_order, which can be verified against its index above

#finding indexes
order_index = raw_order.index
#print(order_index) #used to check if len of DataFrame is correct
num_orders = len(raw_order)
print("The total number of orders created is:",num_orders)

#ii)the total value of the orders created is the sum of every object in the final_price column

#listing the columns of the raw_order DataFrame
order_columns = raw_order.columns
#print(order_columns) #used to check column names if needed

#appending final_price column to a list
final_price = raw_order['final_price']

#finding the total value by summing all elements in the final_price object
print("The total value of all orders is:", sum(final_price))
```

```python
#QUESTION 2: WHICH IS THE MOST POPULAR PRODUCT OFFERED IN THE CATEGORY OF ELECTRONICS?

#i)to find the most popular product, seperate all objects belonging to the electronics category into a DataFrame and sort in descending order by value

sorted_raw_product = raw_product.set_index('category') #setting the index of the raw_product DataFrame to 'category' 
electronics_df = sorted_raw_product.loc['electronics'] #creating a DataFrame for all 'electronics' products
a = electronics_df.sort_values(by = 'Unnamed: 1', ascending = False) #sort electronics_df DataFrame by number ordered in descending order

#printing results (dataframe a is arranged in descending order, max value)
print("The two most popular products offered in the category of electronics have product IDs:",a.iloc[0]['product_id'],"and",a.iloc[1]['product_id'])
print("Both products have",a.iloc[0]['Unnamed: 1'],"orders each")
```

```python
#QUESTION 7: WHAT IS THE AVG DISCOUNT RATE PER PRODUCT CATEGORY

#the discount rate is found by dividing the orignal price by the purchase price, subtracting 1 and multiplying by 100

#i)first sort the products categorys by product_id
sorted_products = raw_product.set_index('category')
books_df = sorted_products.loc['books']
books = books_df['product_id']

electronics_df = sorted_products.loc['electronics']
electronics = electronics_df['product_id']

kitchen_df = sorted_products.loc['kitchen']
kitchen = kitchen_df['product_id']

sports_df = sorted_products.loc['sports']
sports = sports_df['product_id']

#ii)find the average discount rate for each category

#books
raw_order_sorted = raw_order.set_index('product_id')
books_cost = raw_order_sorted.loc[books]

books_orig = books_cost['orginal_price']
books_purchase = books_cost['purchase_price']

avg_books_orig = sum(books_orig)
avg_books_purchase = sum(books_purchase)

books_discount = ((avg_books_orig/avg_books_purchase)-1)*100
print("Book sales have an average discount of",books_discount,'percent')

#electronics
raw_order_sorted = raw_order.set_index('product_id')
electronics_cost = raw_order_sorted.loc[electronics]

electronics_orig = electronics_cost['orginal_price']
electronics_purchase = electronics_cost['purchase_price']

avg_electronics_orig = sum(electronics_orig)
avg_electronics_purchase = sum(electronics_purchase)

electronics_discount = ((avg_electronics_orig/avg_electronics_purchase)-1)*100
print("Electronics sales have an average discount of",electronics_discount,'percent')

#kitchen
raw_order_sorted = raw_order.set_index('product_id')
kitchen_cost = raw_order_sorted.loc[kitchen]

kitchen_orig = kitchen_cost['orginal_price']
kitchen_purchase = kitchen_cost['purchase_price']

avg_kitchen_orig = sum(kitchen_orig)
avg_kitchen_purchase = sum(kitchen_purchase)

kitchen_discount = ((avg_kitchen_orig/avg_kitchen_purchase)-1)*100
print("Kitchen sales have an average discount of",kitchen_discount,'percent')

#sports
raw_order_sorted = raw_order.set_index('product_id')
sports_cost = raw_order_sorted.loc[sports]

sports_orig = sports_cost['orginal_price']
sports_purchase = sports_cost['purchase_price']

avg_sports_orig = sum(sports_orig)
avg_sports_purchase = sum(sports_purchase)

sports_discount = ((avg_sports_orig/avg_sports_purchase)-1)*100
print("Sports sales have an average discount of",sports_discount,'percent')
```
